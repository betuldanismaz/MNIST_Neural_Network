```python
# Loading and preparing the dataset
# Training, validation, and test data are being separated.
training_data, validation_data, test_data = mnist_loader.load_data_wrapper()

# Initializing the neural network model
# Input: 784 (28x28 pixels), Hidden Layer: 30 neurons, Output: 10 (digits)
net = network.Network([784, 30, 10])

# Training the network using Stochastic Gradient Descent (SGD)
# Epochs: 30, Mini-batch: 10, Learning Rate (eta): 3.0
print("--- Basic Model Training Starting ---")
net.SGD(training_data, 30, 10, 3.0, test_data=test_data)
```

```python
import network
import mnist_loader

print("--- Preparing Dataset (Converting to List) ---")
# Convert data to a list for Python 3 compatibility.
# This prevents the data from being "exhausted" and allows testing at each epoch.
training_data, validation_data, test_data = mnist_loader.load_data_wrapper()
training_data = list(training_data)
test_data = list(test_data)

print("--- Experiment A: Training Starts with Batch Size 100 ---")

# 1. Reset the Network (Restart)
net_expA = network.Network([784, 30, 10])

# 2. Start Training
# Parameters: (Data, Epochs=30, Batch=100, Eta=3.0)
net_expA.SGD(training_data, 30, 100, 3.0, test_data=test_data)

```

```python
import network
import mnist_loader

# --- DATA PREPARATION ---
# Ensure data is loaded as lists to prevent iterator exhaustion during testing.
try:
    # Check if data is already a list
    len(training_data)
except (NameError, TypeError):
    print("Loading data and converting to lists...")
    tr_d, va_d, te_d = mnist_loader.load_data_wrapper()
    training_data = list(tr_d)
    validation_data = list(va_d)
    test_data = list(te_d)

# --- EXPERIMENT B: Deep Architecture (2 Hidden Layers) ---
# Change: Added a second hidden layer with 30 neurons.
# Architecture: Input(784) -> Hidden1(30) -> Hidden2(30) -> Output(10)
# Goal: Analyze how increased depth affects learning and accuracy.

print("--- Experiment B: Training Deep Network [784, 30, 30, 10] ---")

# 1. Initialize the Deep Network
net_expB = network.Network([784, 30, 30, 10])

# 2. Start Training (SGD)
# Parameters: Epochs=30, Mini-Batch=10 (Default), Eta=3.0 (Default)
net_expB.SGD(training_data, 30, 10, 3.0, test_data=test_data)
```

```python
import network
import mnist_loader

# --- DATA CHECK ---
# Ensuring data is loaded as lists to prevent errors.
try:
    len(training_data)
except (NameError, TypeError):
    print("Reloading data...")
    tr_d, va_d, te_d = mnist_loader.load_data_wrapper()
    training_data = list(tr_d)
    validation_data = list(va_d)
    test_data = list(te_d)

# --- EXPERIMENT C: Increased Capacity (Neuron Count) ---
# Change: Increased hidden layer neurons from 30 to 100.
# Architecture: [784, 100, 10] (Back to single hidden layer, but wider)
# Batch Size: Reset to 10 (Baseline value)
# Goal: Test if more neurons allow the network to learn more complex patterns.

print("--- Experiment C: Training with 100 Hidden Neurons ---")

# 1. Initialize Network with 100 Hidden Neurons
net_expC = network.Network([784, 100, 10])

# 2. Start Training
# Parameters: Epochs=30, Mini-Batch=10 (Restored to default), Eta=3.0
net_expC.SGD(training_data, 30, 10, 3.0, test_data=test_data)
```
